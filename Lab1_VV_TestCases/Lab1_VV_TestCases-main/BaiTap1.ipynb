```python
import math

def solve_quartic(a, b, c):
    if a == 0 and b == 0 and c == 0:
        return 'Infinite'
    if a == 0 and b == 0:
        return []
    if a == 0:
        y = -c / b
        if y < 0:
            return []
        return [math.sqrt(y), -math.sqrt(y)]
    delta = b*b - 4*a*c
    if delta < 0:
        return []
    y1 = (-b + math.sqrt(delta)) / (2*a)
    y2 = (-b - math.sqrt(delta)) / (2*a)
    res = []
    if y1 >= 0:
        res.extend([math.sqrt(y1), -math.sqrt(y1)])
    if y2 >= 0 and y2 != y1:
        res.extend([math.sqrt(y2), -math.sqrt(y2)])
    return res

print(solve_quartic(1, -5, 4))
```
